```python
from pathlib import Path
import json
from html import escape
from IPython.display import HTML, display

cwd = Path.cwd()

json_dir = cwd / "samples" / "json_manual"

json_files = sorted(json_dir.glob("*.json"))

for json_file in json_files:

    with open(json_file, "r", encoding="utf-8") as f:
        data = json.load(f)

    template = data["narrative"]["template"]

    html_parts = []

    # Document title
    html_parts.append(
        f"""
        <div style="
            background:#ff6b6b;
            color:white;
            padding:12px;
            font-size:22px;
            font-weight:bold;
            border-radius:8px;
            margin-bottom:20px;
        ">
            {escape(template.get("title", ""))}
        </div>
        """
    )

    # Document description, only if available
    template_description = template.get("description", "").strip()

    if template_description:
        html_parts.append(
            f"""
            <div style="
                background:#161b22;
                color:#c9d1d9;
                padding:12px;
                margin-bottom:15px;
                border-left:5px solid #ff6b6b;
                white-space:pre-wrap;
                border-radius:4px;
            ">
                {escape(template_description)}
            </div>
            """
        )

    # Sections
    for section in template.get("section", []):

        html_parts.append(
            f"""
            <div style="
                background:#4dabf7;
                color:white;
                padding:10px;
                font-weight:bold;
                border-radius:6px;
                margin-top:15px;
            ">
                {escape(section.get("title", ""))}
            </div>
            """
        )

        # Section description, only if available
        section_description = section.get("description", "").strip()

        if section_description:
            html_parts.append(
                f"""
                <div style="
                    background:#161b22;
                    color:#c9d1d9;
                    padding:12px;
                    margin-bottom:10px;
                    border-left:5px solid #4dabf7;
                    white-space:pre-wrap;
                    border-radius:4px;
                ">
                    {escape(section_description)}
                </div>
                """
            )

        for question in section.get("question", []):

            question_text = question.get("text", "").strip()

            answer = (
                question
                .get("answer", {})
                .get("json", {})
                .get("answer", "")
            )

            if question_text:

                html_parts.append(
                    f"""
                    <div style="
                        background:#ffd43b;
                        color:black;
                        padding:8px;
                        margin-top:10px;
                        border-radius:4px;
                        font-weight:bold;
                    ">
                        {escape(question_text)}
                    </div>
                    """
                )

            html_parts.append(
                f"""
                <div style="
                    background:#161b22;
                    color:#e6edf3;
                    padding:12px;
                    margin-bottom:10px;
                    border-left:5px solid #51cf66;
                    white-space:pre-wrap;
                    border-radius:4px;
                ">
                    {escape(answer)}
                </div>
                """
            )

    display(
        HTML(
            f"""
            <details open>
                <summary style="
                    font-size:18px;
                    font-weight:bold;
                    padding:10px;
                ">
                    {json_file.stem}
                </summary>

                <div style="
                    background:#0d1117;
                    padding:20px;
                    border-radius:8px;
                    max-height:900px;
                    overflow:auto;
                ">
                    {''.join(html_parts)}
                </div>
            </details>
            """
        )
    )
```
